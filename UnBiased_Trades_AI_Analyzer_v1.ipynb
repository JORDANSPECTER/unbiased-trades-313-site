```python
def full_analysis_with_oil_context(trade, oil_start, oil_end, oil_minutes):
    score = 0

    # Base quality
    if trade.get("quality") == "A":
        score += 3
    elif trade.get("quality") == "B":
        score += 2
    else:
        score += 1

    # Direction + structure
    if trade.get("htf_bias") == "bullish" and trade.get("reaction") in ["break_hold", "reclaim"]:
        score += 2
    if trade.get("htf_bias") == "bearish" and trade.get("reaction") in ["break_hold", "rejection"]:
        score += 2

    if trade.get("mtf_structure") in ["higher_highs", "lower_lows"]:
        score += 2

    if trade.get("position_vs_vwap") == "above" and trade.get("htf_bias") == "bullish":
        score += 2
    if trade.get("position_vs_vwap") == "below" and trade.get("htf_bias") == "bearish":
        score += 2

    # Test count
    if trade.get("test_number") == "first":
        score += 3
    elif trade.get("test_number") == "second":
        score += 1
    elif trade.get("test_number") == "late_cycle_continuation":
        score += 0

    # Confluence
    score += min(len(trade.get("confluence_factors", [])), 3)

    # Best session windows
    if trade.get("session_phase") in ["open", "power_hour"]:
        score += 2

    # Oil context
    oil_shock = detect_oil_shock(oil_start, oil_end, oil_minutes)
    oil_context = oil_to_qqq_context(oil_shock)
    score += oil_context["qqq_bias_boost"]

    # Mistakes + timing penalties
    mgmt_penalty, mgmt_notes = management_penalty(trade)
    score -= mgmt_penalty

    if trade.get("mistake_tag") == "late_entry":
        score -= 1
    if trade.get("mistake_tag") == "chasing":
        score -= 2
    if trade.get("mistake_tag") == "overtrade":
        score -= 2

    # Grade bands
    if score >= 12:
        grade = "A"
        win_probability = 80
    elif score >= 8:
        grade = "B+"
        win_probability = 68
    elif score >= 5:
        grade = "B"
        win_probability = 58
    else:
        grade = "C"
        win_probability = 45

    summary = (
        f"{trade.get('symbol')} {trade.get('direction')} setup | "
        f"reaction={trade.get('reaction')} | "
        f"level={trade.get('level')} | "
        f"session={trade.get('session_phase')} | "
        f"grade={grade}"
    )

    return {
        "symbol": trade.get("symbol"),
        "asset_type": trade.get("asset_type"),
        "score": score,
        "grade": grade,
        "win_probability": win_probability,
        "summary": summary,
        "oil_shock": oil_shock,
        "oil_context": oil_context,
        "management_notes": mgmt_notes,
        "trade": trade
    }
```

```python
def management_penalty(trade):
    penalty = 0
    notes = []

    if trade.get("session_phase") == "late_morning":
        penalty += 1
        notes.append("Late-morning continuation risk")

    if trade.get("rsi_state") == "elevated":
        penalty += 1
        notes.append("RSI elevated")

    if trade.get("macd_state") == "mature_bullish":
        penalty += 1
        notes.append("MACD mature, not fresh")

    if trade.get("bollinger_context") == "near_upper_band":
        penalty += 1
        notes.append("Price extended near upper band")

    if trade.get("trade_management_mistake") == "failed_to_sell_into_resistance":
        penalty += 2
        notes.append("Failed to sell into known resistance cluster")

    return penalty, notes
```

```python
def oil_to_qqq_context(oil_shock_info):
    if not oil_shock_info["oil_shock"]:
        return {
            "qqq_bias_boost": 0,
            "macro_context": "normal",
            "macro_note": "No abnormal oil shock detected."
        }

    if oil_shock_info["shock_direction"] == "down":
        return {
            "qqq_bias_boost": 3 if oil_shock_info["shock_strength"] == "extreme" else 2,
            "macro_context": "bullish_for_qqq",
            "macro_note": "Fast downside oil shock supports upside impulse in QQQ."
        }

    if oil_shock_info["shock_direction"] == "up":
        return {
            "qqq_bias_boost": -3 if oil_shock_info["shock_strength"] == "extreme" else -2,
            "macro_context": "bearish_for_qqq",
            "macro_note": "Fast upside oil shock pressures QQQ."
        }

    return {
        "qqq_bias_boost": 0,
        "macro_context": "neutral",
        "macro_note": "Oil move did not add directional context."
    }
```

```python
def detect_oil_shock(start_price, end_price, minutes, shock_threshold_pct=2.5):
    if start_price <= 0 or minutes <= 0:
        return {
            "oil_shock": False,
            "pct_change": 0.0,
            "velocity_per_min": 0.0,
            "shock_direction": "none",
            "shock_strength": "none"
        }

    pct_change = ((end_price - start_price) / start_price) * 100
    velocity_per_min = pct_change / minutes

    oil_shock = minutes <= 20 and abs(pct_change) >= shock_threshold_pct

    if pct_change < 0:
        shock_direction = "down"
    elif pct_change > 0:
        shock_direction = "up"
    else:
        shock_direction = "flat"

    if abs(pct_change) >= 4:
        shock_strength = "extreme"
    elif abs(pct_change) >= 2.5:
        shock_strength = "strong"
    elif abs(pct_change) >= 1.0:
        shock_strength = "moderate"
    else:
        shock_strength = "light"

    return {
        "oil_shock": oil_shock,
        "pct_change": round(pct_change, 2),
        "velocity_per_min": round(velocity_per_min, 3),
        "shock_direction": shock_direction,
        "shock_strength": shock_strength
    }
```

```python
def macro_bias(oil_trend, macro_event, market_sentiment):
    oil_trend = str(oil_trend).lower()
    macro_event = str(macro_event).lower()
    market_sentiment = str(market_sentiment).lower()

    if oil_trend == "up" and macro_event in ["war", "geopolitical", "conflict"] and market_sentiment == "fear":
        return "bearish_qqq"
    elif oil_trend == "down" and market_sentiment in ["risk_on", "bullish"]:
        return "bullish_qqq"
    else:
        return "neutral"
```

```python
def cleanup_old_journal_defaults(
    default_timeframe="5m",
    default_session="open",
    default_volatility="medium",
    default_move_type="unknown"
):
    global trade_journal
    updated = 0

    for t in trade_journal:
        trade_input = t.get("trade_input", {})

        if not trade_input:
            continue

        symbol = trade_input.get("symbol", t.get("symbol", ""))
        symbol_upper = str(symbol).upper()

        if "asset_type" not in trade_input or not trade_input.get("asset_type"):
            if symbol_upper in ["QQQ", "SPY", "IWM", "DIA"]:
                trade_input["asset_type"] = "index"
            elif symbol_upper:
                trade_input["asset_type"] = "stock"
            else:
                trade_input["asset_type"] = "unknown"
            updated += 1

        if "timeframe" not in trade_input or not trade_input.get("timeframe"):
            trade_input["timeframe"] = default_timeframe
            updated += 1

        if "session" not in trade_input or not trade_input.get("session"):
            trade_input["session"] = default_session
            updated += 1

        if "volatility" not in trade_input or not trade_input.get("volatility"):
            trade_input["volatility"] = default_volatility
            updated += 1

        if "move_type" not in trade_input or not trade_input.get("move_type"):
            trade_input["move_type"] = default_move_type
            updated += 1

        if "htf_bias" not in trade_input or not trade_input.get("htf_bias"):
            trade_input["htf_bias"] = "unknown"
            updated += 1

        if "mtf_structure" not in trade_input or not trade_input.get("mtf_structure"):
            trade_input["mtf_structure"] = "unknown"
            updated += 1

        if "market" not in trade_input or not trade_input.get("market"):
            trade_input["market"] = "unknown"
            updated += 1

        if "test" not in trade_input or not trade_input.get("test"):
            trade_input["test"] = "unknown"
            updated += 1

        if "level" not in trade_input or not trade_input.get("level"):
            trade_input["level"] = "unknown"
            updated += 1

        if "reaction" not in trade_input or not trade_input.get("reaction"):
            trade_input["reaction"] = "unknown"
            updated += 1

        if "vwap" not in trade_input or not trade_input.get("vwap"):
            trade_input["vwap"] = "unknown"
            updated += 1

        if "volume" not in trade_input or not trade_input.get("volume"):
            trade_input["volume"] = "unknown"
            updated += 1

        if "key_level" not in trade_input or not trade_input.get("key_level"):
            trade_input["key_level"] = t.get("key_level", "unknown")
            updated += 1

        t["trade_input"] = trade_input
        t["pattern_signature"] = pattern_signature(trade_input)

    import json
    with open("trade_journal.json", "w") as f:
        json.dump(trade_journal, f, indent=4)

    print(f"✅ Cleanup complete. Fields updated: {updated}")
```

```python
def analyze_structured_trade(data):
    score = 0
    reasons = []
    risks = []

    bias = "NO TRADE"
    setup_type = "None"
    grade = "C"
    confidence = 0

    level = data.get("level", "").lower()
    reaction = data.get("reaction", "").lower()
    vwap = data.get("vwap", "").lower()
    rsi = data.get("rsi", 50)
    volume = data.get("volume", "").lower()
    market = data.get("market", "").lower()
    test = str(data.get("test", "")).lower()
    key_level = data.get("key_level", "User-defined")
    htf_bias = data.get("htf_bias", "").lower()
    mtf_structure = data.get("mtf_structure", "").lower()
    session = data.get("session", "").lower()
    volatility = data.get("volatility", "").lower()
    move_type = data.get("move_type", "").lower()

    no_trade_words = ["chop", "sideways", "indecision", "news", "extended", "unclear"]
    if market in no_trade_words or vwap == "flip" or level == "unclear" or volume == "low":
        return {
            "bias": "NO TRADE",
            "setup_type": "No Trade",
            "grade": "C",
            "key_level": key_level,
            "entry_trigger": "None",
            "invalidation": "None",
            "targets": "None",
            "confidence": 20,
            "reasons": ["Conditions are not clean enough"],
            "risks": ["Chop / weak structure / poor confirmation"],
            "final_decision": "NO TRADE"
        }

    # Core setup classification
    if level == "resistance" and reaction == "rejection":
        score += 2
        bias = "PUT"
        setup_type = "Rejection"
        reasons.append("Rejection at resistance")

    elif level == "resistance" and reaction == "break_hold":
        score += 3
        bias = "CALL"
        setup_type = "Break and Hold"
        reasons.append("Break and hold above resistance")

    elif level == "support" and reaction == "retest_hold":
        score += 2
        bias = "CALL"
        setup_type = "Retest Hold"
        reasons.append("Retest held at support")

    elif level == "resistance" and reaction == "lower_high":
        score += 2
        bias = "PUT"
        setup_type = "Failed Bounce / Lower High"
        reasons.append("Lower high rejection at resistance")

    # VWAP
    if vwap == "below":
        score += 2
        reasons.append("Bearish VWAP control")
    elif vwap == "above":
        score += 2
        reasons.append("Bullish VWAP control")

    # RSI
    if rsi < 50:
        score += 1
        reasons.append("Bearish momentum")
    elif rsi > 50:
        score += 1
        reasons.append("Bullish momentum")

    # Volume
    if volume == "high":
        score += 2
        reasons.append("Volume confirmation")
    else:
        risks.append("No strong volume confirmation")

    # Test quality
    if test == "first":
        score += 1
        reasons.append("First test of level")
    elif test == "second":
        risks.append("Second test is weaker than first")
    elif test == "third":
        risks.append("Third test may be exhausted")

    # Structure alignment
    if htf_bias == "bullish":
        score += 1
        reasons.append("Higher timeframe bullish bias")
    elif htf_bias == "bearish" and bias == "CALL":
        risks.append("Counter-trend to higher timeframe bias")

    if mtf_structure in ["higher_highs", "higher_lows"] and bias == "CALL":
        score += 1
        reasons.append("Bullish multi-timeframe structure")
    elif mtf_structure in ["lower_highs", "lower_lows"] and bias == "CALL":
        risks.append("Weak multi-timeframe structure for CALL")

    # Session / volatility / move type logic
    if session in ["open", "power_hour"] and bias == "CALL":
        score += 1
        reasons.append("Momentum-friendly session")

    if volatility == "high" and bias == "CALL" and setup_type == "Break and Hold":
        score += 1
        reasons.append("High-volatility breakout environment")

    if move_type in ["expansion breakout", "breakout squeeze"] and setup_type == "Break and Hold":
        score += 2
        reasons.append("Expansion breakout / squeeze behavior")

    # Grade
    if score >= 10:
        grade = "A+"
    elif score >= 7:
        grade = "B"
    else:
        grade = "C"

    confidence = min(100, score * 10)

    # Entry / invalidation / targets
    if bias == "CALL" and setup_type == "Break and Hold":
        entry_trigger = "Break above resistance holds with continuation"
        invalidation = "Loss of breakout level / loss of VWAP"
        targets = "Next resistance / extension target"
    elif bias == "CALL":
        entry_trigger = "Retest bounce confirms"
        invalidation = "Loss of support / retest low"
        targets = "Next resistance / next key level"
    elif bias == "PUT":
        entry_trigger = "Rejection confirms or lower high fails"
        invalidation = "Break above rejection high / reclaimed level"
        targets = "Next support / next key level"
    else:
        entry_trigger = "None"
        invalidation = "None"
        targets = "None"

    if grade == "C":
        final_decision = "NO TRADE"
    else:
        final_decision = bias

    return {
        "bias": bias,
        "setup_type": setup_type,
        "grade": grade,
        "key_level": key_level,
        "entry_trigger": entry_trigger,
        "invalidation": invalidation,
        "targets": targets,
        "confidence": confidence,
        "reasons": reasons,
        "risks": risks,
        "final_decision": final_decision
    }
```

```python
def log_trade(
    symbol,
    bias,
    setup_type,
    grade,
    key_level,
    adjusted_confidence,
    position_size,
    result,
    pnl,
    notes="",
    trade_input=None
):
    import json
    from datetime import datetime

    global trade_journal

    entry = {
        "timestamp": datetime.now().isoformat(timespec="seconds"),
        "symbol": symbol,
        "bias": bias,
        "setup_type": setup_type,
        "grade": grade,
        "key_level": key_level,
        "adjusted_confidence": adjusted_confidence,
        "position_size": position_size,
        "result": result,
        "pnl": pnl,
        "notes": notes,
        "trade_input": trade_input if trade_input else {},
        "pattern_signature": pattern_signature(trade_input) if trade_input else ""
    }

    trade_journal.append(entry)

    with open("trade_journal.json", "w") as f:
        json.dump(trade_journal, f, indent=4)

    print("✅ Trade logged and saved")
    return entry
```

```python
def learned_pattern_stats():
    stats = {}

    for t in trade_journal:
        # SAFE pulls (prevents errors)
        asset = t.get("trade_input", {}).get("asset_type", "unknown")
        timeframe = t.get("trade_input", {}).get("timeframe", "unknown")
        htf = t.get("trade_input", {}).get("htf_bias", "unknown")
        structure = t.get("trade_input", {}).get("mtf_structure", "unknown")
        session = t.get("trade_input", {}).get("session", "unknown")
        volatility = t.get("trade_input", {}).get("volatility", "unknown")

        reaction = t.get("trade_input", {}).get("reaction", "unknown")
        vwap = t.get("trade_input", {}).get("vwap", "unknown")
        volume = t.get("trade_input", {}).get("volume", "unknown")
        market = t.get("trade_input", {}).get("market", "unknown")
        test = t.get("trade_input", {}).get("test", "unknown")
        level = t.get("trade_input", {}).get("level", "unknown")

        move = t.get("trade_input", {}).get("move_type", "unknown")

        # BUILD SIGNATURE
        sig = f"asset={asset} | timeframe={timeframe} | htf={htf} | structure={structure} | session={session} | volatility={volatility} | move_type={move} | reaction={reaction} | vwap={vwap} | volume={volume} | market={market} | test={test} | level={level}"

        if sig not in stats:
            stats[sig] = {
                "wins": 0,
                "losses": 0,
                "pnl": 0.0,
                "sample_size": 0
            }

        result = t.get("result")

        if result == "win":
            stats[sig]["wins"] += 1
        elif result == "loss":
            stats[sig]["losses"] += 1

        stats[sig]["pnl"] += float(t.get("pnl", 0))
        stats[sig]["sample_size"] += 1

    # calculate win rate
    for sig, s in stats.items():
        total = s["wins"] + s["losses"]
        s["win_rate"] = round((s["wins"] / total) * 100, 2) if total > 0 else 0.0

    return stats
```

```python
def adjust_confidence(result, pattern):
    category = pattern["category"]

    performance = track_performance()
    category_win_rate = performance.get(category, {}).get("win_rate", 0)

    setup_stats = learned_setup_stats()
    setup_type = result.get("setup_type", "Unknown")
    setup_win_rate = setup_stats.get(setup_type, {}).get("win_rate", 0)
    setup_sample_size = setup_stats.get(setup_type, {}).get("sample_size", 0)

    trade_input = result.get("trade_input", {})
    sig = pattern_signature(trade_input) if trade_input else None
    pattern_stats = learned_pattern_stats()
    deep_pattern = pattern_stats.get(sig, {})
    deep_pattern_win_rate = deep_pattern.get("win_rate", 0)
    deep_pattern_sample_size = deep_pattern.get("sample_size", 0)

    adjusted_confidence = result["confidence"]

    # Category performance
    if category_win_rate >= 70:
        adjusted_confidence += 10
    elif category_win_rate > 50:
        adjusted_confidence += 0
    elif category_win_rate > 0:
        adjusted_confidence -= 15
    else:
        adjusted_confidence -= 25

    # Setup performance
    if setup_sample_size >= 3:
        if setup_win_rate >= 70:
            adjusted_confidence += 10
        elif setup_win_rate <= 40:
            adjusted_confidence -= 10

    # Deep pattern performance
    if deep_pattern_sample_size >= 2:
        if deep_pattern_win_rate >= 70:
            adjusted_confidence += 10
        elif deep_pattern_win_rate <= 40:
            adjusted_confidence -= 10

    # Priority window boost
    trade_window = trade_input.get("trade_window", "unknown")

    if trade_window in ["priority_morning", "priority_power"]:
        adjusted_confidence += 8
    elif trade_window == "priority_midday":
        adjusted_confidence += 4
    elif trade_window == "off_hours":
        adjusted_confidence -= 10

    # Execution discipline
    execution = trade_input.get("execution_discipline", "unknown")
    if execution == "high":
        adjusted_confidence += 4
    elif execution == "low":
        adjusted_confidence -= 8

    # Mistakes
    mistake = trade_input.get("mistake_tag", "none")
    if mistake != "none":
        adjusted_confidence -= 12

    # Environment preference
    environment = trade_input.get("environment", "unknown")
    if environment == "trend_day" and result.get("setup_type") in ["Break and Hold", "Retest Hold"]:
        adjusted_confidence += 5
    elif environment == "range_day" and result.get("setup_type") == "Break and Hold":
        adjusted_confidence -= 8

    # Clamp
    adjusted_confidence = max(0, min(100, adjusted_confidence))

    print("\n=== ADJUSTED CONFIDENCE ===")
    print("Base Confidence:", result["confidence"])
    print("Category:", category)
    print("Category Win Rate:", category_win_rate)
    print("Setup Type:", setup_type)
    print("Setup Win Rate:", setup_win_rate)
    print("Setup Sample Size:", setup_sample_size)
    print("Deep Pattern Signature:", sig)
    print("Deep Pattern Win Rate:", deep_pattern_win_rate)
    print("Deep Pattern Sample Size:", deep_pattern_sample_size)
    print("Trade Window:", trade_window)
    print("Execution Discipline:", execution)
    print("Environment:", environment)
    print("Mistake Tag:", mistake)
    print("Adjusted Confidence:", adjusted_confidence)

    return adjusted_confidence
```

```python
def journal_summary():
    total_trades = len(trade_journal)
    wins = sum(1 for t in trade_journal if t["result"] == "win")
    losses = sum(1 for t in trade_journal if t["result"] == "loss")
    breakevens = sum(1 for t in trade_journal if t["result"] == "breakeven")
    total_pnl = sum(float(t["pnl"]) for t in trade_journal)

    resolved = wins + losses
    win_rate = round((wins / resolved) * 100, 2) if resolved > 0 else 0.0

    print("\n=== JOURNAL SUMMARY ===")
    print("Total Trades:", total_trades)
    print("Wins:", wins)
    print("Losses:", losses)
    print("Breakevens:", breakevens)
    print("Win Rate:", win_rate, "%")
    print("Total PnL:", total_pnl)

    return {
        "total_trades": total_trades,
        "wins": wins,
        "losses": losses,
        "breakevens": breakevens,
        "win_rate": win_rate,
        "total_pnl": total_pnl
    }
```

```python
def position_size(confidence, grade=None, trade_input=None):
    if grade == "A+":
        base_risk = 3
    elif grade == "A":
        base_risk = 2
    elif grade == "B":
        base_risk = 1
    else:
        base_risk = 0.5

    trade_window = "unknown"
    execution = "unknown"
    mistake = "none"

    if trade_input:
        trade_window = trade_input.get("trade_window", "unknown")
        execution = trade_input.get("execution_discipline", "unknown")
        mistake = trade_input.get("mistake_tag", "none")

    # Window sizing adjustment
    if trade_window in ["priority_morning", "priority_power"]:
        base_risk += 0.5
    elif trade_window == "off_hours":
        base_risk -= 0.5

    # Execution adjustment
    if execution == "high":
        base_risk += 0.25
    elif execution == "low":
        base_risk -= 0.5

    # Mistake penalty
    if mistake != "none":
        base_risk -= 0.75

    base_risk = max(0.25, base_risk)

    if confidence >= 90:
        size = "FULL SIZE"
    elif confidence >= 75:
        size = "HALF SIZE"
    elif confidence >= 60:
        size = "QUARTER SIZE"
    else:
        size = "NO TRADE"

    print("\n=== POSITION SIZING ===")
    print("Grade:", grade)
    print("Confidence:", confidence)
    print("Trade Window:", trade_window)
    print("Execution Discipline:", execution)
    print("Mistake Tag:", mistake)
    print("Risk Per Trade:", f"{base_risk}%")
    print("Recommended Size:", size)

    return size
```

```python
def adjust_confidence(result, pattern):
    category = pattern["category"]

    performance = track_performance()

    win_rate = performance.get(category, {}).get("win_rate", 0)

    adjusted_confidence = result["confidence"]

    # Adjust based on performance
    if win_rate >= 70:
        adjusted_confidence += 10
    elif win_rate >= 50:
        adjusted_confidence += 0
    elif win_rate > 0:
        adjusted_confidence -= 15
    else:
        adjusted_confidence -= 25

    # Clamp between 0–100
    adjusted_confidence = max(0, min(100, adjusted_confidence))

    print("\n=== ADJUSTED CONFIDENCE ===")
    print("Base Confidence:", result["confidence"])
    print("Win Rate:", win_rate)
    print("Adjusted Confidence:", adjusted_confidence)

    return adjusted_confidence
```

```python
def track_performance():
    print("\n=== PERFORMANCE TRACKING ===")

    performance = {}

    for category, setups in saved_trades.items():
        wins = 0
        losses = 0

        for name, trade in setups.items():
            result = trade.get("result")

            if result == "win":
                wins += 1
            elif result == "loss":
                losses += 1

        total = wins + losses

        if total > 0:
            win_rate = round((wins / total) * 100, 2)
        else:
            win_rate = 0

        performance[category] = {
            "wins": wins,
            "losses": losses,
            "win_rate": win_rate
        }

        print(f"\n{category}")
        print("Wins:", wins)
        print("Losses:", losses)
        print("Win Rate:", win_rate, "%")

    return performance
```

```python
def best_pattern_match(trade_input):
    best_name = None
    best_category = None
    best_score = -1

    keys_to_compare = [
        "asset_type",
        "level",
        "reaction",
        "vwap",
        "volume",
        "market",
        "test",
        "timeframe",
        "htf_bias",
        "mtf_structure",
        "session",
        "volatility"
    ]

    for category, setups in saved_trades.items():
        for name, saved in setups.items():
            match_score = 0

            for key in keys_to_compare:
                if isinstance(saved, dict) and trade_input.get(key) == saved.get(key):
                    match_score += 1

            if match_score > best_score:
                best_score = match_score
                best_name = name
                best_category = category

    return {
        "name": best_name,
        "category": best_category,
        "match_score": best_score
    }
```

```python
def analyze_structured_trade(data):
    score = 0
    reasons = []
    risks = []

    bias = "NO TRADE"
    setup_type = "None"
    grade = "C"
    confidence = 0

    level = data.get("level", "").lower()
    reaction = data.get("reaction", "").lower()
    vwap = data.get("vwap", "").lower()
    rsi = data.get("rsi", 50)
    volume = data.get("volume", "").lower()
    market = data.get("market", "").lower()
    test = str(data.get("test", "")).lower()
    key_level = data.get("key_level", "User-defined")

    # No-trade conditions
    no_trade_words = ["chop", "sideways", "indecision", "news", "extended", "unclear"]
    if market in no_trade_words or vwap == "flip" or level == "unclear" or volume == "low":
        return {
            "bias": "NO TRADE",
            "setup_type": "No Trade",
            "grade": "C",
            "key_level": key_level,
            "entry_trigger": "None",
            "invalidation": "None",
            "targets": "None",
            "confidence": 20,
            "reasons": ["Conditions are not clean enough"],
            "risks": ["Chop / weak structure / poor confirmation"],
            "final_decision": "NO TRADE"
        }

    # Level + reaction
    if level == "resistance" and reaction == "rejection":
        score += 2
        bias = "PUT"
        setup_type = "Rejection"
        reasons.append("Rejection at resistance")

    elif level == "resistance" and reaction == "break_hold":
        score += 2
        bias = "CALL"
        setup_type = "Break and Hold"
        reasons.append("Break and hold above resistance")

    elif level == "support" and reaction == "retest_hold":
        score += 2
        bias = "CALL"
        setup_type = "Retest Hold"
        reasons.append("Retest held at support")

    elif level == "resistance" and reaction == "lower_high":
        score += 2
        bias = "PUT"
        setup_type = "Failed Bounce / Lower High"
        reasons.append("Lower high rejection at resistance")

    # VWAP
    if vwap == "below":
        score += 2
        reasons.append("Bearish VWAP control")
    elif vwap == "above":
        score += 2
        reasons.append("Bullish VWAP control")

    # RSI
    if rsi < 50:
        score += 1
        reasons.append("Bearish momentum")
    elif rsi > 50:
        score += 1
        reasons.append("Bullish momentum")

    # Volume
    if volume == "high":
        score += 1
        reasons.append("Volume confirmation")
    elif volume != "high":
        risks.append("No strong volume confirmation")

    # Test number
    if test == "first":
        score += 1
        reasons.append("First test of level")
    elif test == "third":
        risks.append("Third test may be weaker / more likely to break")

    # Grade
    if score >= 6:
        grade = "A+"
    elif score >= 4:
        grade = "B"
    else:
        grade = "C"

    confidence = min(100, score * 15)

    # Entry / invalidation / targets
    if bias == "CALL":
        entry_trigger = "Break holds or retest bounce confirms"
        invalidation = "Loss of breakout level / retest low"
        targets = "Next resistance / next key level"
    elif bias == "PUT":
        entry_trigger = "Rejection candle confirms or lower high fails"
        invalidation = "Break above rejection high / reclaimed level"
        targets = "Next support / next key level"
    else:
        entry_trigger = "None"
        invalidation = "None"
        targets = "None"

    # Final decision
    if grade == "C":
        final_decision = "NO TRADE"
    else:
        final_decision = bias

    return {
        "bias": bias,
        "setup_type": setup_type,
        "grade": grade,
        "key_level": key_level,
        "entry_trigger": entry_trigger,
        "invalidation": invalidation,
        "targets": targets,
        "confidence": confidence,
        "reasons": reasons,
        "risks": risks,
        "final_decision": final_decision
    }
```

```python
def analyze_trade(setup):
    setup = setup.lower()

    score = 0
    reasons = []
    risks = []

    # --- NO TRADE CONDITIONS ---
    no_trade_conditions = [
        "midday", "chop", "sideways", "vwap flipping", "indecision",
        "no volume", "no level", "extended", "news"
    ]

    for cond in no_trade_conditions:
        if cond in setup:
            return {
                "bias": "NEUTRAL",
                "setup_type": "NONE",
                "grade": "N/A",
                "key_level": "unclear",
                "entry_trigger": "none",
                "invalidation": "none",
                "targets": "none",
                "confidence": 20,
                "reasons": ["Market conditions not tradable"],
                "risks": ["Chop / no structure"],
                "final_decision": "NO TRADE"
            }

    # --- VWAP ---
    if "above vwap" in setup or "vwap support" in setup:
        score += 2
        reasons.append("Bullish VWAP control")

    if "below vwap" in setup or "vwap reject" in setup:
        score += 2
        reasons.append("Bearish VWAP control")

    # --- RSI ---
    if "rsi above 50" in setup or "rsi 55" in setup or "rsi 60" in setup:
        score += 1
        reasons.append("Bullish momentum")

    if "rsi below 50" in setup or "rsi 40" in setup:
        score += 1
        reasons.append("Bearish momentum")

    # --- PRICE ACTION ---
    if "break and hold" in setup or "breakout" in setup:
        score += 2
        reasons.append("Break and hold behavior")

    if "retest hold" in setup:
        score += 2
        reasons.append("Successful retest")

    if "rejection" in setup or "wick" in setup:
        score += 2
        reasons.append("Rejection at level")

    if "lower high" in setup:
        score += 1
        reasons.append("Weak structure")

    # --- VOLUME ---
    if "volume" in setup:
        score += 1
        reasons.append("Volume confirmation")

    # --- DETERMINE BIAS ---
    if "rejection" in setup or "lower high" in setup:
        bias = "PUT"
        setup_type = "Rejection"

    elif "break" in setup or "hold" in setup:
        bias = "CALL"
        setup_type = "Break and Hold"

    else:
        bias = "NEUTRAL"
        setup_type = "Unclear"

    # --- GRADE ---
    if score >= 6:
        grade = "A+"
    elif score >= 4:
        grade = "B"
    else:
        grade = "C"

    # --- CONFIDENCE ---
    confidence = min(100, score * 15)

    # --- RISK FLAGS ---
    if score < 4:
        risks.append("Weak confluence")

    if "extended" in setup:
        risks.append("Late entry risk")

    # --- FINAL DECISION ---
    if grade == "C":
        decision = "NO TRADE"
    elif bias == "CALL":
        decision = "CALL"
    elif bias == "PUT":
        decision = "PUT"
    else:
        decision = "NO TRADE"

    return {
        "bias": bias,
        "setup_type": setup_type,
        "grade": grade,
        "key_level": "User-defined",
        "entry_trigger": "Confirmation candle / retest",
        "invalidation": "Break of level / opposite structure",
        "targets": "Next key level",
        "confidence": confidence,
        "reasons": reasons,
        "risks": risks,
        "final_decision": decision
    }
```

```python
def analyze_setup(user_input):
    user_input = user_input.lower()

    score = 0
    reasons = []

    # VWAP logic
    if "vwap reject" in user_input or "below vwap" in user_input:
        score += 2
        reasons.append("VWAP bearish control")

    if "vwap support" in user_input or "above vwap" in user_input:
        score += 2
        reasons.append("VWAP bullish control")

    # RSI logic
    if "rsi below 50" in user_input:
        score += 1
        reasons.append("Bearish momentum")

    if "rsi above 50" in user_input:
        score += 1
        reasons.append("Bullish momentum")

    # Price action
    if "reject" in user_input:
        score += 2
        reasons.append("Rejection at level")

    if "break" in user_input and "hold" in user_input:
        score += 2
        reasons.append("Break and hold")

    # Market condition
    if "sideways" in user_input or "chop" in user_input:
        return "❌ No Trade — Market is in chop (low probability)"

    # Determine bias
    if "reject" in user_input:
        bias = "PUT"
    elif "break" in user_input:
        bias = "CALL"
    else:
        bias = "NEUTRAL"

    # Grade setup
    if score >= 5:
        grade = "A+"
    elif score >= 3:
        grade = "B"
    else:
        grade = "C"

    return f"""
Setup: {bias}
Grade: {grade}
Confidence Score: {score}/7

Reasons:
- {"\n- ".join(reasons)}
"""
```
