```python
def score_night_before_setup(plan):
    score = 0
    notes = []

    # Higher timeframe bias
    if plan["higher_timeframe_thesis"]["bias"] in ["bullish", "bullish but extended"]:
        score += 2
        notes.append("Higher timeframe trend supports upside structure.")

    # Institutional anchor
    if plan["institutional_logic"]["primary_anchor"]:
        score += 3
        notes.append("Clear institutional anchor exists.")

    # Defined support and resistance
    if len(plan["core_levels"]["support_zones"]) > 0 and len(plan["core_levels"]["resistance_zones"]) > 0:
        score += 2
        notes.append("Plan has clean map of support and resistance.")

    # No-trade zone awareness
    if plan.get("no_trade_zone"):
        score += 2
        notes.append("Plan includes a no-trade zone, which improves discipline.")

    # Extension penalty
    if "extended" in plan["higher_timeframe_thesis"]["bias"]:
        score -= 2
        notes.append("Price is extended, reducing reward-to-risk for fresh longs.")

    # Mixed momentum penalty
    macd_read = plan["indicator_state"]["macd"]["read"].lower()
    rsi_read = plan["indicator_state"]["rsi"]["read"].lower()
    if "slowing" in macd_read or "flattening" in macd_read:
        score -= 1
        notes.append("Momentum is slowing.")
    if "cooling" in rsi_read:
        score -= 1
        notes.append("RSI is elevated but cooling.")

    # Final grade
    if score >= 8:
        grade = "A-"
    elif score >= 6:
        grade = "B+"
    elif score >= 4:
        grade = "B"
    else:
        grade = "C"

    return {
        "score": score,
        "grade": grade,
        "notes": notes
    }


result = score_night_before_setup(night_before_plan)
print(result)
```

```python
def format_discord_alert(decision: BotDecision, context: MarketContext) -> str:
    emoji = {
        "BULL": "🟢",
        "BEAR": "🔴",
        "CHOP": "🟡"
    }.get(decision.regime, "⚪")

    grade_emoji = {
        "A+": "💎",
        "A": "🔥",
        "B+": "✅",
        "B": "⚠️",
        "AVOID": "⛔"
    }.get(decision.grade, "⚪")

    lines = [
        f"{emoji} **{context.symbol} Regime Update**",
        f"**Regime:** {decision.regime}",
        f"**Allowed Direction:** {decision.allowed_direction}",
        f"**Setup:** {decision.setup or 'None'}",
        f"**Confidence:** {decision.confidence}",
        f"**Grade:** {grade_emoji} {decision.grade}",
        f"**Grade Score:** {decision.grade_score}",
        f"**Entry Valid:** {'YES' if decision.entry_valid else 'NO'}",
    ]

    if decision.stop_level is not None:
        lines.append(f"**Stop Reference:** {decision.stop_level:.2f}")
    if decision.target_level is not None:
        lines.append(f"**Target Reference:** {decision.target_level:.2f}")

    lines.append("")
    lines.append("**Reasoning:**")

    for item in decision.reasoning[:12]:
        lines.append(f"- {item}")

    return "\n".join(lines)
```

```python
def make_decision(
    context: MarketContext,
    structure: StructureState,
    candle_time: datetime,
) -> BotDecision:
    regime_data = detect_regime(context, structure, candle_time)
    setup = detect_setup(regime_data["regime"], structure)
    entry = confirm_entry(regime_data["regime"], setup, context, structure)
    confidence = confidence_label(regime_data, entry["entry_valid"])

    grade_data = grade_setup(
        regime_data=regime_data,
        setup=setup,
        entry_data=entry,
        context=context,
        structure=structure,
        candle_time=candle_time,
    )

    reasoning = []
    reasoning.extend(regime_data["notes"])

    if setup:
        reasoning.append(f"Detected setup: {setup}.")
    else:
        reasoning.append("No valid setup detected.")

    reasoning.extend(entry["reasons"])
    reasoning.extend(grade_data["grade_notes"])

    return BotDecision(
        regime=regime_data["regime"],
        allowed_direction=regime_data["allowed_direction"],
        setup=setup,
        confidence=confidence,
        grade=grade_data["grade"],
        grade_score=grade_data["score"],
        entry_valid=entry["entry_valid"],
        stop_level=entry["stop_level"],
        target_level=entry["target_level"],
        reasoning=reasoning,
    )
```

```python
from typing import Dict, Optional
from datetime import datetime


def grade_setup(
    regime_data: Dict[str, object],
    setup: Optional[str],
    entry_data: Dict[str, object],
    context: MarketContext,
    structure: StructureState,
    candle_time: datetime,
) -> Dict[str, object]:
    score = 0
    notes = []

    bull_score = regime_data.get("bull_score", 0)
    bear_score = regime_data.get("bear_score", 0)
    regime = regime_data.get("regime", "CHOP")

    # 1. Regime strength
    spread = abs(bull_score - bear_score)
    if spread >= 5:
        score += 3
        notes.append("Strong regime separation.")
    elif spread >= 3:
        score += 2
        notes.append("Good regime separation.")
    elif spread >= 2:
        score += 1
        notes.append("Moderate regime separation.")
    else:
        notes.append("Weak regime separation.")

    # 2. Valid setup present
    if setup is not None:
        score += 2
        notes.append("Valid setup detected.")
    else:
        notes.append("No clean setup detected.")

    # 3. Entry confirmation
    if entry_data.get("entry_valid", False):
        score += 3
        notes.append("Entry confirmation is valid.")
    else:
        notes.append("Entry confirmation is not valid.")

    # 4. VWAP alignment
    if regime == "BULL" and structure.above_vwap:
        score += 1
        notes.append("Bull setup aligned above VWAP.")
    elif regime == "BEAR" and structure.below_vwap:
        score += 1
        notes.append("Bear setup aligned below VWAP.")
    else:
        notes.append("VWAP alignment is weak or mixed.")

    # 5. Volume / reaction quality
    if structure.breakout_with_volume:
        score += 1
        notes.append("Breakout volume supports move.")
    if structure.retest_hold:
        score += 1
        notes.append("Retest hold improves setup quality.")
    if structure.rejection_at_level:
        score += 1
        notes.append("Rejection at level improves bearish quality.")
    if structure.failed_bounce:
        score += 1
        notes.append("Failed bounce improves bearish quality.")

    # 6. Macro alignment
    if regime == "BULL" and context.oil_trend in ["falling", "stabilizing"]:
        score += 1
        notes.append("Macro/oil conditions support upside.")
    elif regime == "BEAR" and context.oil_change_dollars >= 2 and context.oil_trend == "rising":
        score += 1
        notes.append("Macro/oil conditions support downside.")
    else:
        notes.append("Macro alignment is neutral or mixed.")

    # 7. Institutional time window
    if is_priority_window(candle_time):
        score += 1
        notes.append("Signal is in a priority institutional window.")
    else:
        notes.append("Signal is outside priority window.")

    # 8. Chop penalty
    if regime == "CHOP":
        score -= 3
        notes.append("Chop regime penalty applied.")

    # 9. Direct conflict penalty
    if regime == "BULL" and structure.below_vwap:
        score -= 2
        notes.append("Conflict: bull regime but below VWAP.")
    if regime == "BEAR" and structure.above_vwap:
        score -= 2
        notes.append("Conflict: bear regime but above VWAP.")

    # Final grade mapping
    if not entry_data.get("entry_valid", False) or setup is None:
        if score >= 5:
            grade = "B"
        else:
            grade = "AVOID"
    else:
        if score >= 11:
            grade = "A+"
        elif score >= 9:
            grade = "A"
        elif score >= 7:
            grade = "B+"
        elif score >= 5:
            grade = "B"
        else:
            grade = "AVOID"

    return {
        "grade": grade,
        "score": score,
        "grade_notes": notes,
    }
```

```python
from dataclasses import dataclass
from typing import List, Optional, Dict
from datetime import datetime


# =========================
# DATA STRUCTURES
# =========================

@dataclass
class MarketContext:
    symbol: str
    current_price: float
    vwap: float
    rsi: float
    oil_change_dollars: float
    oil_trend: str
    macro_headline_bias: str
    big_print_levels: List[float]
    prior_day_high: float
    prior_day_low: float
    premarket_high: float
    premarket_low: float
    key_supports: List[float]
    key_resistances: List[float]


@dataclass
class StructureState:
    higher_lows: bool
    lower_highs: bool
    above_vwap: bool
    below_vwap: bool
    vwap_reclaimed: bool
    vwap_rejected: bool
    breakout_with_volume: bool
    rejection_at_level: bool
    retest_hold: bool
    failed_bounce: bool
    liquidity_grab_reversal: bool


@dataclass
@dataclass
class BotDecision:
    regime: str
    allowed_direction: str
    setup: Optional[str]
    confidence: str
    grade: str
    grade_score: int
    entry_valid: bool
    stop_level: Optional[float]
    target_level: Optional[float]
    reasoning: List[str]


# =========================
# HELPER
# =========================

def nearest_level(price: float, levels: List[float]) -> Optional[float]:
    if not levels:
        return None
    return min(levels, key=lambda x: abs(x - price))


# =========================
# REGIME DETECTION
# =========================

def detect_regime(
    context: MarketContext,
    structure: StructureState,
    candle_time: datetime,
) -> Dict[str, object]:
    bull_score = 0
    bear_score = 0
    notes = []

    if structure.above_vwap:
        bull_score += 2
        notes.append("Price is above VWAP.")
    if structure.below_vwap:
        bear_score += 2
        notes.append("Price is below VWAP.")

    if structure.higher_lows:
        bull_score += 2
        notes.append("Higher lows support bullish structure.")
    if structure.lower_highs:
        bear_score += 2
        notes.append("Lower highs support bearish structure.")

    if structure.vwap_reclaimed:
        bull_score += 2
        notes.append("VWAP reclaim supports bullish continuation.")
    if structure.vwap_rejected:
        bear_score += 2
        notes.append("VWAP rejection supports bearish continuation.")

    if context.rsi >= 55 and context.rsi < 75:
        bull_score += 1
        notes.append("RSI supports bullish momentum.")
    elif context.rsi <= 45:
        bear_score += 1
        notes.append("RSI supports bearish pressure.")
    elif context.rsi >= 75:
        notes.append("RSI is elevated; bullish but extended.")

    if context.oil_change_dollars >= 2:
        bear_score += 1
        notes.append("Oil strength adds pressure to equities.")
    if context.oil_trend == "stabilizing":
        bull_score += 1
        notes.append("Oil stabilizing may relieve equity pressure.")

    if context.macro_headline_bias == "bullish":
        bull_score += 1
        notes.append("Macro headlines support bullish bias.")
    elif context.macro_headline_bias == "bearish":
        bear_score += 1
        notes.append("Macro headlines support bearish bias.")

    if bull_score - bear_score >= 2:
        regime = "BULL"
        allowed_direction = "LONG"
    elif bear_score - bull_score >= 2:
        regime = "BEAR"
        allowed_direction = "SHORT"
    else:
        regime = "CHOP"
        allowed_direction = "NEUTRAL"

    return {
        "regime": regime,
        "allowed_direction": allowed_direction,
        "bull_score": bull_score,
        "bear_score": bear_score,
        "notes": notes,
    }


# =========================
# SETUP DETECTION
# =========================

def detect_setup(regime: str, structure: StructureState) -> Optional[str]:
    if regime == "BULL":
        if structure.vwap_reclaimed and structure.retest_hold:
            return "VWAP_RECLAIM_RETEST"
        if structure.breakout_with_volume:
            return "BREAKOUT_WITH_VOLUME"
        if structure.liquidity_grab_reversal:
            return "LIQUIDITY_GRAB_REVERSAL"

    if regime == "BEAR":
        if structure.vwap_rejected and structure.rejection_at_level:
            return "VWAP_REJECTION"
        if structure.failed_bounce:
            return "FAILED_BOUNCE"
        if structure.liquidity_grab_reversal:
            return "LIQUIDITY_GRAB_REVERSAL"

    return None


# =========================
# ENTRY CONFIRMATION
# =========================

def confirm_entry(
    regime: str,
    setup: Optional[str],
    context: MarketContext,
    structure: StructureState,
) -> Dict[str, object]:
    reasons = []
    entry_valid = False
    stop_level = None
    target_level = None

    all_supports = [
        context.prior_day_low,
        context.premarket_low,
        *context.key_supports,
        *context.big_print_levels,
    ]

    all_resistances = [
        context.prior_day_high,
        context.premarket_high,
        *context.key_resistances,
        *context.big_print_levels,
    ]

    if regime == "BULL" and setup:
        if structure.above_vwap and (structure.retest_hold or structure.breakout_with_volume):
            entry_valid = True
            reasons.append("Bull regime aligns with bullish setup.")
            reasons.append("Price is holding above VWAP.")

            if structure.retest_hold:
                reasons.append("Retest hold confirms buyers defended the level.")
            if structure.breakout_with_volume:
                reasons.append("Breakout volume confirms participation.")

            stop_level = nearest_level(context.current_price, all_supports)
            target_level = nearest_level(context.current_price, all_resistances)

    elif regime == "BEAR" and setup:
        if structure.below_vwap and (structure.rejection_at_level or structure.failed_bounce):
            entry_valid = True
            reasons.append("Bear regime aligns with bearish setup.")
            reasons.append("Price is staying below VWAP.")

            if structure.rejection_at_level:
                reasons.append("Rejection at resistance confirms sellers are active.")
            if structure.failed_bounce:
                reasons.append("Failed bounce confirms weak buyer control.")

            stop_level = nearest_level(context.current_price, all_resistances)
            target_level = nearest_level(context.current_price, all_supports)

    else:
        reasons.append("No clean entry because regime and setup are not aligned.")

    return {
        "entry_valid": entry_valid,
        "reasons": reasons,
        "stop_level": stop_level,
        "target_level": target_level,
    }


# =========================
# CONFIDENCE LABEL
# =========================

def confidence_label(regime_data: Dict[str, object], entry_valid: bool) -> str:
    bull_score = regime_data.get("bull_score", 0)
    bear_score = regime_data.get("bear_score", 0)
    spread = abs(bull_score - bear_score)

    if not entry_valid:
        return "LOW"
    if spread >= 4:
        return "HIGH"
    if spread >= 2:
        return "MEDIUM"
    return "LOW"


# =========================
# MAIN DECISION ENGINE
# =========================

def make_decision(
    context: MarketContext,
    structure: StructureState,
    candle_time: datetime,
) -> BotDecision:

    regime_data = detect_regime(context, structure, candle_time)
    setup = detect_setup(regime_data["regime"], structure)
    entry = confirm_entry(regime_data["regime"], setup, context, structure)
    confidence = confidence_label(regime_data, entry["entry_valid"])

    # ✅ ADD THIS
    grading = grade_setup(
        regime_data=regime_data,
        setup=setup,
        entry_data=entry,
        context=context,
        structure=structure,
        candle_time=candle_time,
    )

    reasoning = []
    reasoning.extend(regime_data.get("notes", []))
    reasoning.extend(entry.get("reasons", []))

    # ✅ THIS MUST BE INSIDE FUNCTION
    return BotDecision(
        regime=regime_data["regime"],
        allowed_direction=regime_data["allowed_direction"],
        setup=setup,
        confidence=confidence,
        grade=grading["grade"],
        grade_score=grading["score"],
        entry_valid=entry["entry_valid"],
        stop_level=entry["stop_level"],
        target_level=entry["target_level"],
        reasoning=reasoning + grading["grade_notes"],
    )

# =========================
# DISCORD FORMATTER
# =========================

def format_discord_alert(decision: BotDecision, context: MarketContext) -> str:
    emoji = {
        "BULL": "🟢",
        "BEAR": "🔴",
        "CHOP": "🟡",
    }.get(decision.regime, "⚪")

    lines = [
        f"{emoji} **{context.symbol} Regime Update**",
        f"**Regime:** {decision.regime}",
        f"**Allowed Direction:** {decision.allowed_direction}",
        f"**Setup:** {decision.setup or 'None'}",
        f"**Confidence:** {decision.confidence}",
        f"**Entry Valid:** {'YES' if decision.entry_valid else 'NO'}",
    ]

    if decision.stop_level is not None:
        lines.append(f"**Stop Reference:** {decision.stop_level:.2f}")

    if decision.target_level is not None:
        lines.append(f"**Target Reference:** {decision.target_level:.2f}")

    lines.append("")
    lines.append("**Reasoning:**")

    for item in decision.reasoning[:10]:
        lines.append(f"- {item}")

    return "\n".join(lines)


# =========================
# TEST EXAMPLE
# =========================

context = MarketContext(
    symbol="QQQ",
    current_price=613.85,
    vwap=613.20,
    rsi=61.5,
    oil_change_dollars=2.35,
    oil_trend="stabilizing",
    macro_headline_bias="neutral",
    big_print_levels=[613.75],
    prior_day_high=614.40,
    prior_day_low=608.90,
    premarket_high=613.95,
    premarket_low=611.80,
    key_supports=[612.50, 611.80],
    key_resistances=[614.40, 615.20],
)

structure = StructureState(
    higher_lows=True,
    lower_highs=False,
    above_vwap=True,
    below_vwap=False,
    vwap_reclaimed=True,
    vwap_rejected=False,
    breakout_with_volume=True,
    rejection_at_level=False,
    retest_hold=True,
    failed_bounce=False,
    liquidity_grab_reversal=False,
)

decision = make_decision(
    context=context,
    structure=structure,
    candle_time=datetime.now(),
)

print(format_discord_alert(decision, context))
```

```python
def detect_regime(context: MarketContext, structure: StructureState, ts: datetime) -> Dict[str, object]:
    bull_score = 0
    bear_score = 0
    notes = []

    # VWAP
    if structure.above_vwap:
        bull_score += 2
        notes.append("Price is above VWAP.")
    if structure.below_vwap:
        bear_score += 2
        notes.append("Price is below VWAP.")
    if structure.vwap_reclaimed:
        bull_score += 2
        notes.append("VWAP reclaim detected.")
    if structure.vwap_rejected:
        bear_score += 2
        notes.append("VWAP rejection detected.")

    # Structure
    if structure.higher_lows:
        bull_score += 2
        notes.append("Higher lows support bullish structure.")
    if structure.lower_highs:
        bear_score += 2
        notes.append("Lower highs support bearish structure.")

    # Setup behavior
    if structure.breakout_with_volume:
        bull_score += 1
        notes.append("Breakout with volume supports continuation.")
    if structure.rejection_at_level:
        bear_score += 1
        notes.append("Rejection at level supports downside pressure.")
    if structure.retest_hold:
        bull_score += 1
        notes.append("Retest hold supports bullish continuation.")
    if structure.failed_bounce:
        bear_score += 1
        notes.append("Failed bounce supports bearish continuation.")

    # Macro overlay
    if context.oil_change_dollars >= 2 and context.oil_trend == "rising":
        bear_score += 2
        notes.append("Oil rising sharply adds pressure to the market.")
    elif context.oil_trend == "stabilizing":
        bull_score += 1
        notes.append("Oil stabilizing may allow market relief.")
    elif context.oil_trend == "falling":
        bull_score += 1
        notes.append("Oil falling reduces macro pressure.")

    if context.macro_headline_bias == "bullish":
        bull_score += 1
        notes.append("Macro headline bias is bullish.")
    elif context.macro_headline_bias == "bearish":
        bear_score += 1
        notes.append("Macro headline bias is bearish.")

    if is_priority_window(ts):
        notes.append("Signal occurred during a priority institutional time window.")

    # Final regime
    if bull_score >= bear_score + 2:
        regime = "BULL"
        allowed_direction = "CALLS_ONLY"
    elif bear_score >= bull_score + 2:
        regime = "BEAR"
        allowed_direction = "PUTS_ONLY"
    else:
        regime = "CHOP"
        allowed_direction = "SCALPS_OR_AVOID"

    notes.append(f"Bull score: {bull_score}")
    notes.append(f"Bear score: {bear_score}")
    notes.append(f"Regime classified as {regime}.")

    return {
        "regime": regime,
        "allowed_direction": allowed_direction,
        "notes": notes,
        "bull_score": bull_score,
        "bear_score": bear_score,
    }
```

```python
def is_priority_window(ts: datetime) -> bool:
    """
    Priority windows:
    9:30-11:00 and 15:00-16:15
    """
    current = ts.time()
    morning_start = time(9, 30)
    morning_end = time(11, 0)
    power_start = time(15, 0)
    power_end = time(16, 15)

    return (morning_start <= current <= morning_end) or (power_start <= current <= power_end)


def nearest_level(price: float, levels: List[float]) -> Optional[float]:
    clean_levels = [lvl for lvl in levels if lvl is not None]
    if not clean_levels:
      return None
    return min(clean_levels, key=lambda x: abs(x - price))
```

```python
def update_trade_lifecycle(live_df):
    global trade_state

    row = live_df.iloc[0]
    price = row["price"]

    # -------------------------
    # ENTER TRADE
    # -------------------------
    if not trade_state["active"] and row["action"] in ["FULL_SIZE_ENTRY", "NORMAL_ENTRY"]:
        trade_state["active"] = True
        trade_state["entry_price"] = price
        trade_state["stop_loss"] = row["stop_loss"]
        trade_state["target"] = row["target"]
        trade_state["r_multiple"] = row["r_multiple"]

        print("🚀 Trade Opened")

    # -------------------------
    # MANAGE ACTIVE TRADE
    # -------------------------
    elif trade_state["active"]:
        entry = trade_state["entry_price"]
        stop = trade_state["stop_loss"]

        if isinstance(stop, (int, float)):
            risk = entry - stop
            move = price - entry
            r_gain = move / risk if risk != 0 else 0

            print(f"📊 Live R Gain: {round(r_gain, 2)}R")

        # Stop hit
        if isinstance(stop, (int, float)) and price <= stop:
            print("❌ Stopped Out")
            trade_state["active"] = False

        # Target hit
        if isinstance(trade_state["target"], (int, float)) and price >= trade_state["target"]:
            print("💰 Target Hit")
            trade_state["active"] = False

    return trade_state
```

```python
def trailing_stop_logic(row):
    price = row["price"]
    entry = row.get("entry_price", price)
    stop = row["stop_loss"]

    # Only apply to active trades
    if row["action"] not in ["FULL_SIZE_ENTRY", "NORMAL_ENTRY", "SMALL_SIZE_ENTRY"]:
        return stop

    risk = entry - stop if isinstance(stop, (int, float)) else None

    if risk is None or risk == 0:
        return stop

    # Calculate R gained
    move = price - entry
    r_gain = move / risk

    # 1R → breakeven
    if r_gain >= 1:
        stop = entry

    # 2R → trail lock
    if r_gain >= 2:
        stop = round(price - 0.5, 2)

    return stop
```

```python
def trade_action(row):
    patterns = row.get("patterns", [])
    decision = row["decision"]

    # hard profit logic first
    if should_force_take_profit(row):
        return "TAKE_PROFIT"

    # entry logic
    entry = entry_signal(row)
    if entry is not None:
        return entry

    # negative state
    if decision == "SKIP_OR_EXIT_FAST":
        return "EXIT_OR_AVOID"

    if "late_entry_risk" in patterns and "momentum_extended" in patterns:
        return "REDUCE_OR_TIGHTEN_STOP"

    return "WAIT"
```

```python
def full_analysis_with_oil_context(trade, oil_start, oil_end, oil_minutes):
    score = 0

    # Base quality
    if trade.get("quality") == "A":
        score += 3
    elif trade.get("quality") == "B":
        score += 2
    else:
        score += 1

    # Direction + structure
    if trade.get("htf_bias") == "bullish" and trade.get("reaction") in ["break_hold", "reclaim"]:
        score += 2
    if trade.get("htf_bias") == "bearish" and trade.get("reaction") in ["break_hold", "rejection"]:
        score += 2

    if trade.get("mtf_structure") in ["higher_highs", "lower_lows"]:
        score += 2

    if trade.get("position_vs_vwap") == "above" and trade.get("htf_bias") == "bullish":
        score += 2
    if trade.get("position_vs_vwap") == "below" and trade.get("htf_bias") == "bearish":
        score += 2

    # Test count
    if trade.get("test_number") == "first":
        score += 3
    elif trade.get("test_number") == "second":
        score += 1
    elif trade.get("test_number") == "late_cycle_continuation":
        score += 0

    # Confluence
    score += min(len(trade.get("confluence_factors", [])), 3)

    # Best session windows
    if trade.get("session_phase") in ["open", "power_hour"]:
        score += 2

    # Oil context
    oil_shock = detect_oil_shock(oil_start, oil_end, oil_minutes)
    oil_context = oil_to_qqq_context(oil_shock)
    score += oil_context["qqq_bias_boost"]

    # Mistakes + timing penalties
    mgmt_penalty, mgmt_notes = management_penalty(trade)
    score -= mgmt_penalty

    if trade.get("mistake_tag") == "late_entry":
        score -= 1
    if trade.get("mistake_tag") == "chasing":
        score -= 2
    if trade.get("mistake_tag") == "overtrade":
        score -= 2

    # Grade bands
    if score >= 12:
        grade = "A"
        win_probability = 80
    elif score >= 8:
        grade = "B+"
        win_probability = 68
    elif score >= 5:
        grade = "B"
        win_probability = 58
    else:
        grade = "C"
        win_probability = 45

    summary = (
        f"{trade.get('symbol')} {trade.get('direction')} setup | "
        f"reaction={trade.get('reaction')} | "
        f"level={trade.get('level')} | "
        f"session={trade.get('session_phase')} | "
        f"grade={grade}"
    )

    return {
        "symbol": trade.get("symbol"),
        "asset_type": trade.get("asset_type"),
        "score": score,
        "grade": grade,
        "win_probability": win_probability,
        "summary": summary,
        "oil_shock": oil_shock,
        "oil_context": oil_context,
        "management_notes": mgmt_notes,
        "trade": trade
    }
```

```python
def management_penalty(trade):
    penalty = 0
    notes = []

    if trade.get("session_phase") == "late_morning":
        penalty += 1
        notes.append("Late-morning continuation risk")

    if trade.get("rsi_state") == "elevated":
        penalty += 1
        notes.append("RSI elevated")

    if trade.get("macd_state") == "mature_bullish":
        penalty += 1
        notes.append("MACD mature, not fresh")

    if trade.get("bollinger_context") == "near_upper_band":
        penalty += 1
        notes.append("Price extended near upper band")

    if trade.get("trade_management_mistake") == "failed_to_sell_into_resistance":
        penalty += 2
        notes.append("Failed to sell into known resistance cluster")

    return penalty, notes
```

```python
def oil_to_qqq_context(oil_shock_info):
    if not oil_shock_info["oil_shock"]:
        return {
            "qqq_bias_boost": 0,
            "macro_context": "normal",
            "macro_note": "No abnormal oil shock detected."
        }

    if oil_shock_info["shock_direction"] == "down":
        return {
            "qqq_bias_boost": 3 if oil_shock_info["shock_strength"] == "extreme" else 2,
            "macro_context": "bullish_for_qqq",
            "macro_note": "Fast downside oil shock supports upside impulse in QQQ."
        }

    if oil_shock_info["shock_direction"] == "up":
        return {
            "qqq_bias_boost": -3 if oil_shock_info["shock_strength"] == "extreme" else -2,
            "macro_context": "bearish_for_qqq",
            "macro_note": "Fast upside oil shock pressures QQQ."
        }

    return {
        "qqq_bias_boost": 0,
        "macro_context": "neutral",
        "macro_note": "Oil move did not add directional context."
    }
```

```python
def detect_oil_shock(start_price, end_price, minutes, shock_threshold_pct=2.5):
    if start_price <= 0 or minutes <= 0:
        return {
            "oil_shock": False,
            "pct_change": 0.0,
            "velocity_per_min": 0.0,
            "shock_direction": "none",
            "shock_strength": "none"
        }

    pct_change = ((end_price - start_price) / start_price) * 100
    velocity_per_min = pct_change / minutes

    oil_shock = minutes <= 20 and abs(pct_change) >= shock_threshold_pct

    if pct_change < 0:
        shock_direction = "down"
    elif pct_change > 0:
        shock_direction = "up"
    else:
        shock_direction = "flat"

    if abs(pct_change) >= 4:
        shock_strength = "extreme"
    elif abs(pct_change) >= 2.5:
        shock_strength = "strong"
    elif abs(pct_change) >= 1.0:
        shock_strength = "moderate"
    else:
        shock_strength = "light"

    return {
        "oil_shock": oil_shock,
        "pct_change": round(pct_change, 2),
        "velocity_per_min": round(velocity_per_min, 3),
        "shock_direction": shock_direction,
        "shock_strength": shock_strength
    }
```

```python
def macro_bias(oil_trend, macro_event, market_sentiment):
    oil_trend = str(oil_trend).lower()
    macro_event = str(macro_event).lower()
    market_sentiment = str(market_sentiment).lower()

    if oil_trend == "up" and macro_event in ["war", "geopolitical", "conflict"] and market_sentiment == "fear":
        return "bearish_qqq"
    elif oil_trend == "down" and market_sentiment in ["risk_on", "bullish"]:
        return "bullish_qqq"
    else:
        return "neutral"
```

```python
def cleanup_old_journal_defaults(
    default_timeframe="5m",
    default_session="open",
    default_volatility="medium",
    default_move_type="unknown"
):
    global trade_journal
    updated = 0

    for t in trade_journal:
        trade_input = t.get("trade_input", {})

        if not trade_input:
            continue

        symbol = trade_input.get("symbol", t.get("symbol", ""))
        symbol_upper = str(symbol).upper()

        if "asset_type" not in trade_input or not trade_input.get("asset_type"):
            if symbol_upper in ["QQQ", "SPY", "IWM", "DIA"]:
                trade_input["asset_type"] = "index"
            elif symbol_upper:
                trade_input["asset_type"] = "stock"
            else:
                trade_input["asset_type"] = "unknown"
            updated += 1

        if "timeframe" not in trade_input or not trade_input.get("timeframe"):
            trade_input["timeframe"] = default_timeframe
            updated += 1

        if "session" not in trade_input or not trade_input.get("session"):
            trade_input["session"] = default_session
            updated += 1

        if "volatility" not in trade_input or not trade_input.get("volatility"):
            trade_input["volatility"] = default_volatility
            updated += 1

        if "move_type" not in trade_input or not trade_input.get("move_type"):
            trade_input["move_type"] = default_move_type
            updated += 1

        if "htf_bias" not in trade_input or not trade_input.get("htf_bias"):
            trade_input["htf_bias"] = "unknown"
            updated += 1

        if "mtf_structure" not in trade_input or not trade_input.get("mtf_structure"):
            trade_input["mtf_structure"] = "unknown"
            updated += 1

        if "market" not in trade_input or not trade_input.get("market"):
            trade_input["market"] = "unknown"
            updated += 1

        if "test" not in trade_input or not trade_input.get("test"):
            trade_input["test"] = "unknown"
            updated += 1

        if "level" not in trade_input or not trade_input.get("level"):
            trade_input["level"] = "unknown"
            updated += 1

        if "reaction" not in trade_input or not trade_input.get("reaction"):
            trade_input["reaction"] = "unknown"
            updated += 1

        if "vwap" not in trade_input or not trade_input.get("vwap"):
            trade_input["vwap"] = "unknown"
            updated += 1

        if "volume" not in trade_input or not trade_input.get("volume"):
            trade_input["volume"] = "unknown"
            updated += 1

        if "key_level" not in trade_input or not trade_input.get("key_level"):
            trade_input["key_level"] = t.get("key_level", "unknown")
            updated += 1

        t["trade_input"] = trade_input
        t["pattern_signature"] = pattern_signature(trade_input)

    import json
    with open("trade_journal.json", "w") as f:
        json.dump(trade_journal, f, indent=4)

    print(f"✅ Cleanup complete. Fields updated: {updated}")
```

```python
def analyze_structured_trade(data):
    score = 0
    reasons = []
    risks = []

    bias = "NO TRADE"
    setup_type = "None"
    grade = "C"
    confidence = 0

    level = data.get("level", "").lower()
    reaction = data.get("reaction", "").lower()
    vwap = data.get("vwap", "").lower()
    rsi = data.get("rsi", 50)
    volume = data.get("volume", "").lower()
    market = data.get("market", "").lower()
    test = str(data.get("test", "")).lower()
    key_level = data.get("key_level", "User-defined")
    htf_bias = data.get("htf_bias", "").lower()
    mtf_structure = data.get("mtf_structure", "").lower()
    session = data.get("session", "").lower()
    volatility = data.get("volatility", "").lower()
    move_type = data.get("move_type", "").lower()

    no_trade_words = ["chop", "sideways", "indecision", "news", "extended", "unclear"]
    if market in no_trade_words or vwap == "flip" or level == "unclear" or volume == "low":
        return {
            "bias": "NO TRADE",
            "setup_type": "No Trade",
            "grade": "C",
            "key_level": key_level,
            "entry_trigger": "None",
            "invalidation": "None",
            "targets": "None",
            "confidence": 20,
            "reasons": ["Conditions are not clean enough"],
            "risks": ["Chop / weak structure / poor confirmation"],
            "final_decision": "NO TRADE"
        }

    # Core setup classification
    if level == "resistance" and reaction == "rejection":
        score += 2
        bias = "PUT"
        setup_type = "Rejection"
        reasons.append("Rejection at resistance")

    elif level == "resistance" and reaction == "break_hold":
        score += 3
        bias = "CALL"
        setup_type = "Break and Hold"
        reasons.append("Break and hold above resistance")

    elif level == "support" and reaction == "retest_hold":
        score += 2
        bias = "CALL"
        setup_type = "Retest Hold"
        reasons.append("Retest held at support")

    elif level == "resistance" and reaction == "lower_high":
        score += 2
        bias = "PUT"
        setup_type = "Failed Bounce / Lower High"
        reasons.append("Lower high rejection at resistance")

    # VWAP
    if vwap == "below":
        score += 2
        reasons.append("Bearish VWAP control")
    elif vwap == "above":
        score += 2
        reasons.append("Bullish VWAP control")

    # RSI
    if rsi < 50:
        score += 1
        reasons.append("Bearish momentum")
    elif rsi > 50:
        score += 1
        reasons.append("Bullish momentum")

    # Volume
    if volume == "high":
        score += 2
        reasons.append("Volume confirmation")
    else:
        risks.append("No strong volume confirmation")

    # Test quality
    if test == "first":
        score += 1
        reasons.append("First test of level")
    elif test == "second":
        risks.append("Second test is weaker than first")
    elif test == "third":
        risks.append("Third test may be exhausted")

    # Structure alignment
    if htf_bias == "bullish":
        score += 1
        reasons.append("Higher timeframe bullish bias")
    elif htf_bias == "bearish" and bias == "CALL":
        risks.append("Counter-trend to higher timeframe bias")

    if mtf_structure in ["higher_highs", "higher_lows"] and bias == "CALL":
        score += 1
        reasons.append("Bullish multi-timeframe structure")
    elif mtf_structure in ["lower_highs", "lower_lows"] and bias == "CALL":
        risks.append("Weak multi-timeframe structure for CALL")

    # Session / volatility / move type logic
    if session in ["open", "power_hour"] and bias == "CALL":
        score += 1
        reasons.append("Momentum-friendly session")

    if volatility == "high" and bias == "CALL" and setup_type == "Break and Hold":
        score += 1
        reasons.append("High-volatility breakout environment")

    if move_type in ["expansion breakout", "breakout squeeze"] and setup_type == "Break and Hold":
        score += 2
        reasons.append("Expansion breakout / squeeze behavior")

    # Grade
    if score >= 10:
        grade = "A+"
    elif score >= 7:
        grade = "B"
    else:
        grade = "C"

    confidence = min(100, score * 10)

    # Entry / invalidation / targets
    if bias == "CALL" and setup_type == "Break and Hold":
        entry_trigger = "Break above resistance holds with continuation"
        invalidation = "Loss of breakout level / loss of VWAP"
        targets = "Next resistance / extension target"
    elif bias == "CALL":
        entry_trigger = "Retest bounce confirms"
        invalidation = "Loss of support / retest low"
        targets = "Next resistance / next key level"
    elif bias == "PUT":
        entry_trigger = "Rejection confirms or lower high fails"
        invalidation = "Break above rejection high / reclaimed level"
        targets = "Next support / next key level"
    else:
        entry_trigger = "None"
        invalidation = "None"
        targets = "None"

    if grade == "C":
        final_decision = "NO TRADE"
    else:
        final_decision = bias

    return {
        "bias": bias,
        "setup_type": setup_type,
        "grade": grade,
        "key_level": key_level,
        "entry_trigger": entry_trigger,
        "invalidation": invalidation,
        "targets": targets,
        "confidence": confidence,
        "reasons": reasons,
        "risks": risks,
        "final_decision": final_decision
    }
```

```python
def log_trade(
    symbol,
    bias,
    setup_type,
    grade,
    key_level,
    adjusted_confidence,
    position_size,
    result,
    pnl,
    notes="",
    trade_input=None
):
    import json
    from datetime import datetime

    global trade_journal

    entry = {
        "timestamp": datetime.now().isoformat(timespec="seconds"),
        "symbol": symbol,
        "bias": bias,
        "setup_type": setup_type,
        "grade": grade,
        "key_level": key_level,
        "adjusted_confidence": adjusted_confidence,
        "position_size": position_size,
        "result": result,
        "pnl": pnl,
        "notes": notes,
        "trade_input": trade_input if trade_input else {},
        "pattern_signature": pattern_signature(trade_input) if trade_input else ""
    }

    trade_journal.append(entry)

    with open("trade_journal.json", "w") as f:
        json.dump(trade_journal, f, indent=4)

    print("✅ Trade logged and saved")
    return entry
```

```python
def log_trade(symbol, setup_type, result, pnl, notes=""):
    import json
    from datetime import datetime

    trade = {
        "symbol": symbol,
        "setup": setup_type,
        "result": result,  # win / loss / breakeven
        "pnl": pnl,
        "notes": notes,
        "timestamp": str(datetime.now())
    }

    try:
        with open("trade_journal.json", "r") as f:
            journal = json.load(f)
    except:
        journal = []

    journal.append(trade)

    with open("trade_journal.json", "w") as f:
        json.dump(journal, f, indent=4)

    print("✅ Trade logged")
```

```python
def learned_pattern_stats():
    stats = {}

    for t in trade_journal:
        # SAFE pulls (prevents errors)
        asset = t.get("trade_input", {}).get("asset_type", "unknown")
        timeframe = t.get("trade_input", {}).get("timeframe", "unknown")
        htf = t.get("trade_input", {}).get("htf_bias", "unknown")
        structure = t.get("trade_input", {}).get("mtf_structure", "unknown")
        session = t.get("trade_input", {}).get("session", "unknown")
        volatility = t.get("trade_input", {}).get("volatility", "unknown")

        reaction = t.get("trade_input", {}).get("reaction", "unknown")
        vwap = t.get("trade_input", {}).get("vwap", "unknown")
        volume = t.get("trade_input", {}).get("volume", "unknown")
        market = t.get("trade_input", {}).get("market", "unknown")
        test = t.get("trade_input", {}).get("test", "unknown")
        level = t.get("trade_input", {}).get("level", "unknown")

        move = t.get("trade_input", {}).get("move_type", "unknown")

        # BUILD SIGNATURE
        sig = f"asset={asset} | timeframe={timeframe} | htf={htf} | structure={structure} | session={session} | volatility={volatility} | move_type={move} | reaction={reaction} | vwap={vwap} | volume={volume} | market={market} | test={test} | level={level}"

        if sig not in stats:
            stats[sig] = {
                "wins": 0,
                "losses": 0,
                "pnl": 0.0,
                "sample_size": 0
            }

        result = t.get("result")

        if result == "win":
            stats[sig]["wins"] += 1
        elif result == "loss":
            stats[sig]["losses"] += 1

        stats[sig]["pnl"] += float(t.get("pnl", 0))
        stats[sig]["sample_size"] += 1

    # calculate win rate
    for sig, s in stats.items():
        total = s["wins"] + s["losses"]
        s["win_rate"] = round((s["wins"] / total) * 100, 2) if total > 0 else 0.0

    return stats
```

```python
def adjust_confidence(result, pattern):
    category = pattern["category"]

    performance = track_performance()
    category_win_rate = performance.get(category, {}).get("win_rate", 0)

    setup_stats = learned_setup_stats()
    setup_type = result.get("setup_type", "Unknown")
    setup_win_rate = setup_stats.get(setup_type, {}).get("win_rate", 0)
    setup_sample_size = setup_stats.get(setup_type, {}).get("sample_size", 0)

    trade_input = result.get("trade_input", {})
    sig = pattern_signature(trade_input) if trade_input else None
    pattern_stats = learned_pattern_stats()
    deep_pattern = pattern_stats.get(sig, {})
    deep_pattern_win_rate = deep_pattern.get("win_rate", 0)
    deep_pattern_sample_size = deep_pattern.get("sample_size", 0)

    adjusted_confidence = result["confidence"]

    # Category performance
    if category_win_rate >= 70:
        adjusted_confidence += 10
    elif category_win_rate > 50:
        adjusted_confidence += 0
    elif category_win_rate > 0:
        adjusted_confidence -= 15
    else:
        adjusted_confidence -= 25

    # Setup performance
    if setup_sample_size >= 3:
        if setup_win_rate >= 70:
            adjusted_confidence += 10
        elif setup_win_rate <= 40:
            adjusted_confidence -= 10

    # Deep pattern performance
    if deep_pattern_sample_size >= 2:
        if deep_pattern_win_rate >= 70:
            adjusted_confidence += 10
        elif deep_pattern_win_rate <= 40:
            adjusted_confidence -= 10

    # Priority window boost
    trade_window = trade_input.get("trade_window", "unknown")

    if trade_window in ["priority_morning", "priority_power"]:
        adjusted_confidence += 8
    elif trade_window == "priority_midday":
        adjusted_confidence += 4
    elif trade_window == "off_hours":
        adjusted_confidence -= 10

    # Execution discipline
    execution = trade_input.get("execution_discipline", "unknown")
    if execution == "high":
        adjusted_confidence += 4
    elif execution == "low":
        adjusted_confidence -= 8

    # Mistakes
    mistake = trade_input.get("mistake_tag", "none")
    if mistake != "none":
        adjusted_confidence -= 12

    # Environment preference
    environment = trade_input.get("environment", "unknown")
    if environment == "trend_day" and result.get("setup_type") in ["Break and Hold", "Retest Hold"]:
        adjusted_confidence += 5
    elif environment == "range_day" and result.get("setup_type") == "Break and Hold":
        adjusted_confidence -= 8

    # Clamp
    adjusted_confidence = max(0, min(100, adjusted_confidence))

    print("\n=== ADJUSTED CONFIDENCE ===")
    print("Base Confidence:", result["confidence"])
    print("Category:", category)
    print("Category Win Rate:", category_win_rate)
    print("Setup Type:", setup_type)
    print("Setup Win Rate:", setup_win_rate)
    print("Setup Sample Size:", setup_sample_size)
    print("Deep Pattern Signature:", sig)
    print("Deep Pattern Win Rate:", deep_pattern_win_rate)
    print("Deep Pattern Sample Size:", deep_pattern_sample_size)
    print("Trade Window:", trade_window)
    print("Execution Discipline:", execution)
    print("Environment:", environment)
    print("Mistake Tag:", mistake)
    print("Adjusted Confidence:", adjusted_confidence)

    return adjusted_confidence
```

```python
def position_size(confidence, grade=None, trade_input=None):
    if grade == "A+":
        base_risk = 3
    elif grade == "A":
        base_risk = 2
    elif grade == "B":
        base_risk = 1
    else:
        base_risk = 0.5

    trade_window = "unknown"
    execution = "unknown"
    mistake = "none"

    if trade_input:
        trade_window = trade_input.get("trade_window", "unknown")
        execution = trade_input.get("execution_discipline", "unknown")
        mistake = trade_input.get("mistake_tag", "none")

    # Window sizing adjustment
    if trade_window in ["priority_morning", "priority_power"]:
        base_risk += 0.5
    elif trade_window == "off_hours":
        base_risk -= 0.5

    # Execution adjustment
    if execution == "high":
        base_risk += 0.25
    elif execution == "low":
        base_risk -= 0.5

    # Mistake penalty
    if mistake != "none":
        base_risk -= 0.75

    base_risk = max(0.25, base_risk)

    if confidence >= 90:
        size = "FULL SIZE"
    elif confidence >= 75:
        size = "HALF SIZE"
    elif confidence >= 60:
        size = "QUARTER SIZE"
    else:
        size = "NO TRADE"

    print("\n=== POSITION SIZING ===")
    print("Grade:", grade)
    print("Confidence:", confidence)
    print("Trade Window:", trade_window)
    print("Execution Discipline:", execution)
    print("Mistake Tag:", mistake)
    print("Risk Per Trade:", f"{base_risk}%")
    print("Recommended Size:", size)

    return size
```

```python
def track_performance():
    print("\n=== PERFORMANCE TRACKING ===")

    performance = {}

    for category, setups in saved_trades.items():
        wins = 0
        losses = 0

        for name, trade in setups.items():
            result = trade.get("result")

            if result == "win":
                wins += 1
            elif result == "loss":
                losses += 1

        total = wins + losses

        if total > 0:
            win_rate = round((wins / total) * 100, 2)
        else:
            win_rate = 0

        performance[category] = {
            "wins": wins,
            "losses": losses,
            "win_rate": win_rate
        }

        print(f"\n{category}")
        print("Wins:", wins)
        print("Losses:", losses)
        print("Win Rate:", win_rate, "%")

    return performance
```

```python
def analyze_structured_trade(data):
    score = 0
    reasons = []
    risks = []

    bias = "NO TRADE"
    setup_type = "None"
    grade = "C"
    confidence = 0

    level = data.get("level", "").lower()
    reaction = data.get("reaction", "").lower()
    vwap = data.get("vwap", "").lower()
    rsi = data.get("rsi", 50)
    volume = data.get("volume", "").lower()
    market = data.get("market", "").lower()
    test = str(data.get("test", "")).lower()
    key_level = data.get("key_level", "User-defined")

    # No-trade conditions
    no_trade_words = ["chop", "sideways", "indecision", "news", "extended", "unclear"]
    if market in no_trade_words or vwap == "flip" or level == "unclear" or volume == "low":
        return {
            "bias": "NO TRADE",
            "setup_type": "No Trade",
            "grade": "C",
            "key_level": key_level,
            "entry_trigger": "None",
            "invalidation": "None",
            "targets": "None",
            "confidence": 20,
            "reasons": ["Conditions are not clean enough"],
            "risks": ["Chop / weak structure / poor confirmation"],
            "final_decision": "NO TRADE"
        }

    # Level + reaction
    if level == "resistance" and reaction == "rejection":
        score += 2
        bias = "PUT"
        setup_type = "Rejection"
        reasons.append("Rejection at resistance")

    elif level == "resistance" and reaction == "break_hold":
        score += 2
        bias = "CALL"
        setup_type = "Break and Hold"
        reasons.append("Break and hold above resistance")

    elif level == "support" and reaction == "retest_hold":
        score += 2
        bias = "CALL"
        setup_type = "Retest Hold"
        reasons.append("Retest held at support")

    elif level == "resistance" and reaction == "lower_high":
        score += 2
        bias = "PUT"
        setup_type = "Failed Bounce / Lower High"
        reasons.append("Lower high rejection at resistance")

    # VWAP
    if vwap == "below":
        score += 2
        reasons.append("Bearish VWAP control")
    elif vwap == "above":
        score += 2
        reasons.append("Bullish VWAP control")

    # RSI
    if rsi < 50:
        score += 1
        reasons.append("Bearish momentum")
    elif rsi > 50:
        score += 1
        reasons.append("Bullish momentum")

    # Volume
    if volume == "high":
        score += 1
        reasons.append("Volume confirmation")
    elif volume != "high":
        risks.append("No strong volume confirmation")

    # Test number
    if test == "first":
        score += 1
        reasons.append("First test of level")
    elif test == "third":
        risks.append("Third test may be weaker / more likely to break")

    # Grade
    if score >= 6:
        grade = "A+"
    elif score >= 4:
        grade = "B"
    else:
        grade = "C"

    confidence = min(100, score * 15)

    # Entry / invalidation / targets
    if bias == "CALL":
        entry_trigger = "Break holds or retest bounce confirms"
        invalidation = "Loss of breakout level / retest low"
        targets = "Next resistance / next key level"
    elif bias == "PUT":
        entry_trigger = "Rejection candle confirms or lower high fails"
        invalidation = "Break above rejection high / reclaimed level"
        targets = "Next support / next key level"
    else:
        entry_trigger = "None"
        invalidation = "None"
        targets = "None"

    # Final decision
    if grade == "C":
        final_decision = "NO TRADE"
    else:
        final_decision = bias

    return {
        "bias": bias,
        "setup_type": setup_type,
        "grade": grade,
        "key_level": key_level,
        "entry_trigger": entry_trigger,
        "invalidation": invalidation,
        "targets": targets,
        "confidence": confidence,
        "reasons": reasons,
        "risks": risks,
        "final_decision": final_decision
    }
```

```python
def analyze_structured_trade(data):
    score = 0
    reasons = []
    risks = []

    # --- LEVEL ---
    if data["level"] == "resistance" and data["reaction"] == "rejection":
        score += 2
        reasons.append("Rejection at resistance")

    if data["level"] == "resistance" and data["reaction"] == "break_hold":
        score += 2
        reasons.append("Break and hold above resistance")

    # --- VWAP ---
    if data["vwap"] == "below":
        score += 2
        reasons.append("Bearish VWAP control")

    elif data["vwap"] == "above":
        score += 2
        reasons.append("Bullish VWAP control")

    # --- RSI ---
    if data["rsi"] < 50:
        score += 1
        reasons.append("Bearish momentum")

    elif data["rsi"] > 50:
        score += 1
        reasons.append("Bullish momentum")

    # --- VOLUME ---
    if data["volume"] == "high":
        score += 1
        reasons.append("Volume confirmation")

    # --- DETERMINE BIAS ---
    if data["reaction"] == "rejection":
        bias = "PUT"
        setup_type = "Rejection"
    elif data["reaction"] == "break_hold":
        bias = "CALL"
        setup_type = "Break and Hold"
    else:
        bias = "NEUTRAL"
        setup_type = "Unclear"

    # --- GRADE ---
    if score >= 6:
        grade = "A+"
    elif score >= 4:
        grade = "B"
    else:
        grade = "C"

    confidence = min(100, score * 15)

    if grade == "C":
        decision = "NO TRADE"
    else:
        decision = bias

    return {
        "bias": bias,
        "setup_type": setup_type,
        "grade": grade,
        "confidence": confidence,
        "reasons": reasons,
        "final_decision": decision
    }
```

```python
def analyze_trade(setup):
    setup = setup.lower()

    score = 0
    reasons = []
    risks = []

    # --- NO TRADE CONDITIONS ---
    no_trade_conditions = [
        "midday", "chop", "sideways", "vwap flipping", "indecision",
        "no volume", "no level", "extended", "news"
    ]

    for cond in no_trade_conditions:
        if cond in setup:
            return {
                "bias": "NEUTRAL",
                "setup_type": "NONE",
                "grade": "N/A",
                "key_level": "unclear",
                "entry_trigger": "none",
                "invalidation": "none",
                "targets": "none",
                "confidence": 20,
                "reasons": ["Market conditions not tradable"],
                "risks": ["Chop / no structure"],
                "final_decision": "NO TRADE"
            }

    # --- VWAP ---
    if "above vwap" in setup or "vwap support" in setup:
        score += 2
        reasons.append("Bullish VWAP control")

    if "below vwap" in setup or "vwap reject" in setup:
        score += 2
        reasons.append("Bearish VWAP control")

    # --- RSI ---
    if "rsi above 50" in setup or "rsi 55" in setup or "rsi 60" in setup:
        score += 1
        reasons.append("Bullish momentum")

    if "rsi below 50" in setup or "rsi 40" in setup:
        score += 1
        reasons.append("Bearish momentum")

    # --- PRICE ACTION ---
    if "break and hold" in setup or "breakout" in setup:
        score += 2
        reasons.append("Break and hold behavior")

    if "retest hold" in setup:
        score += 2
        reasons.append("Successful retest")

    if "rejection" in setup or "wick" in setup:
        score += 2
        reasons.append("Rejection at level")

    if "lower high" in setup:
        score += 1
        reasons.append("Weak structure")

    # --- VOLUME ---
    if "volume" in setup:
        score += 1
        reasons.append("Volume confirmation")

    # --- DETERMINE BIAS ---
    if "rejection" in setup or "lower high" in setup:
        bias = "PUT"
        setup_type = "Rejection"

    elif "break" in setup or "hold" in setup:
        bias = "CALL"
        setup_type = "Break and Hold"

    else:
        bias = "NEUTRAL"
        setup_type = "Unclear"

    # --- GRADE ---
    if score >= 6:
        grade = "A+"
    elif score >= 4:
        grade = "B"
    else:
        grade = "C"

    # --- CONFIDENCE ---
    confidence = min(100, score * 15)

    # --- RISK FLAGS ---
    if score < 4:
        risks.append("Weak confluence")

    if "extended" in setup:
        risks.append("Late entry risk")

    # --- FINAL DECISION ---
    if grade == "C":
        decision = "NO TRADE"
    elif bias == "CALL":
        decision = "CALL"
    elif bias == "PUT":
        decision = "PUT"
    else:
        decision = "NO TRADE"

    return {
        "bias": bias,
        "setup_type": setup_type,
        "grade": grade,
        "key_level": "User-defined",
        "entry_trigger": "Confirmation candle / retest",
        "invalidation": "Break of level / opposite structure",
        "targets": "Next key level",
        "confidence": confidence,
        "reasons": reasons,
        "risks": risks,
        "final_decision": decision
    }
```

```python
def analyze_setup(user_input):
    user_input = user_input.lower()

    score = 0
    reasons = []

    # VWAP logic
    if "vwap reject" in user_input or "below vwap" in user_input:
        score += 2
        reasons.append("VWAP bearish control")

    if "vwap support" in user_input or "above vwap" in user_input:
        score += 2
        reasons.append("VWAP bullish control")

    # RSI logic
    if "rsi below 50" in user_input:
        score += 1
        reasons.append("Bearish momentum")

    if "rsi above 50" in user_input:
        score += 1
        reasons.append("Bullish momentum")

    # Price action
    if "reject" in user_input:
        score += 2
        reasons.append("Rejection at level")

    if "break" in user_input and "hold" in user_input:
        score += 2
        reasons.append("Break and hold")

    # Market condition
    if "sideways" in user_input or "chop" in user_input:
        return "❌ No Trade — Market is in chop (low probability)"

    # Determine bias
    if "reject" in user_input:
        bias = "PUT"
    elif "break" in user_input:
        bias = "CALL"
    else:
        bias = "NEUTRAL"

    # Grade setup
    if score >= 5:
        grade = "A+"
    elif score >= 3:
        grade = "B"
    else:
        grade = "C"

    return f"""
Setup: {bias}
Grade: {grade}
Confidence Score: {score}/7

Reasons:
- {"\n- ".join(reasons)}
"""
```
